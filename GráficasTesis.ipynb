```python
import matplotlib.pyplot as plt

# Datos de asistencia promedio por liga
ligas = ['Bundesliga (Alemania)', 'Premier League (Inglaterra)', 'Serie A (Italia)',
         'Primera División (España)', 'Ligue 1 (Francia)', 'Liga MX (México)',
         'Série A (Brasil)', 'MLS (Estados Unidos)']
asistencia_promedio = [42966, 40301, 29508, 29428, 23677, 22562, 21522, 21034]

# Crear la gráfica
plt.figure(figsize=(13, 6))
plt.barh(ligas, asistencia_promedio, color='skyblue', edgecolor='black')

# Etiqueta del eje X con fuente 12
plt.xlabel('Asistencia Promedio (Número de aficionados)', fontsize=12)

# Ajustar tamaño de las etiquetas de los ejes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Invertir eje Y para que la liga con más asistencia esté arriba
plt.gca().invert_yaxis()

# Agregar líneas de referencia en el eje X
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Añadir anotaciones a las barras con fuente 12 y un ligero desplazamiento
for index, value in enumerate(asistencia_promedio):
    plt.text(value + 100, index, f"{value:,}", fontsize=10, va='center')

# Formatear los números en el eje x con comas
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))

# Ajustar el diseño y mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de asistencia promedio y capacidad promedio por liga
ligas = ['Bundesliga (Alemania)', 'Premier League (Inglaterra)', 'Serie A (Italia)',
         'Primera División (España)', 'Ligue 1 (Francia)', 'Liga MX (México)',
         'Série A (Brasil)', 'MLS (Estados Unidos)']
asistencia_promedio = [42966, 40301, 29508, 29428, 23677, 22562, 21522, 21034]
capacidad_promedio = [46527.17, 40466, 37430.33, 34904.35, 30910.85, 45417, 48122.95, 23803.18]

# Calcular el porcentaje de ocupación
porcentaje_ocupacion = [(asistencia / capacidad) * 100 for asistencia, capacidad in zip(asistencia_promedio, capacidad_promedio)]

# Ordenar los datos por porcentaje de ocupación
datos_ordenados = sorted(zip(ligas, asistencia_promedio, capacidad_promedio, porcentaje_ocupacion), key=lambda x: x[3], reverse=True)
ligas_ordenadas, asistencia_ordenada, capacidad_ordenada, ocupacion_ordenada = zip(*datos_ordenados)

# Crear la gráfica
plt.figure(figsize=(14, 5))
plt.barh(ligas_ordenadas, ocupacion_ordenada, color='skyblue', edgecolor='black', height=0.7)

# Ajustar etiquetas con fuente 12
plt.xlabel('Porcentaje de Ocupación (%)', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Ajustar el límite del eje x
plt.xlim(0, 105)
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Añadir anotaciones a las barras con fuente 10
for index, value in enumerate(ocupacion_ordenada):
    plt.text(value + 0.5, index, f"{value:.2f}%", va='center', fontsize=10)

# Ajustar el diseño y mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
años = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
asistencia = np.array([23754, 25677, 25649, 24130, 22791, 21790, 11780, 11253, 20225, 23525])

# Filtrar los años 2020 y 2021 para la línea de tendencia
años_tendencia = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2022, 2023])
asistencia_tendencia = np.array([23754, 25677, 25649, 24130, 22791, 21790, 20225, 23525])

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
plt.bar(años, asistencia, color='orange', label='Asistencia')

# Calcular la línea de tendencia sin 2020 y 2021
z = np.polyfit(años_tendencia, asistencia_tendencia, 1)
p = np.poly1d(z)

# Dibujar la línea de tendencia
plt.plot(años, p(años), "r--", label='Tendencia (sin 2020 y 2021)')

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Promedio de Asistencia')
plt.title('Promedio de Asistencia Anual de la Liga MX')
plt.ylim(0, 30000)  # Ajustar el límite del eje y
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Asegurar que todos los años se muestren en el eje X
plt.xticks(años)

# Mostrar valores sobre cada barra
for i in range(len(años)):
    plt.text(años[i], asistencia[i] + 500, asistencia[i], ha='center', va='bottom')

# Añadir la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
